```python
import numpy as np
import matplotlib.pyplot as plt
poisson = np.random.poisson(1, 1000)
plt.hist(poisson, bins='auto')
plt.show()
```

```python
uniform = np.random.uniform(0,1, 1000)
plt.hist(uniform, bins='auto')
plt.show()
```

```python
binomial = np.random.binomial(10, .5, 1000)
plt.hist(binomial, bins='auto')
plt.show()
```

```python
binomial2 = np.random.binomial(10, .5, 100000)
normal =  np.random.normal(5, 2, 100000)
fig, axs = plt.subplots(1, 2)
fig.suptitle('binomial vs. normal')
axs[0].hist(binomial2, bins =20)
axs[1].hist(normal, bins = 20)
plt.show()
```
